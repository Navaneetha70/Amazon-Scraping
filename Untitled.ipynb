```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url= "https://www.amazon.com/s?k=playstation+5"
```

```python
#Headers for request
Headers=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36','Accept-language':'en-Us,en;q=0.5'})
```

```python
#HTTP Request
webpage=requests.get(url,headers=Headers)
```

```python
type(webpage.content)
```

```python
#soup object containing all data
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
soup
```

```python
#Fetch links as list of tagobjects
links=soup.find_all("a",attrs={'class':'a-link-normal s-no-outline'})
```

```python
link=links[0].get('href')
```

```python
product_list="https://amazon.com"+link
```

```python
product_list
```

```python
new_webpage =requests.get(product_list, headers=Headers)
```

```python
new_webpage
```

```python
#soup object containing all data
new_soup=BeautifulSoup(new_webpage.content,"html.parser")
```

```python
new_soup
```

```python
new_soup.find("span",attrs={"id":'productTitle'}).text.strip()
```
